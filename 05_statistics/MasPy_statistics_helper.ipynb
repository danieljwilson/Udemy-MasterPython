```python
# import all necessary modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
```

# Download and inspect the data

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

data = pd.read_csv(url, sep=';')
data
```

```python
# describe the data
data.describe()
```

```python
# list number of unique values per column

for col in data.columns:
    u_vals = len(np.unique(data[col]))
    print(f'{col} has {u_vals} unique values')
```

```python
# plot some data
fig,ax = plt.subplots(1,figsize=(17,4))

# data.boxplot()
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)

plt.show()
```

```python
# remove rows with outliers
# based of being more than 3 sd from the mean
data = data[(np.abs(stats.zscore(data)) < 4).all(axis=1)]
```

```python
# some exploration (scatterplot of any two columns)
sns.scatterplot(x=data['residual sugar'], y=data['alcohol'])
plt.show()

# pairwise plots (pick three columns)
sns.pairplot(data, vars=data.columns[:3],
             palette='colorblind',
             kind='reg',
             diag_kind='kde',
             hue='quality'
            )
plt.show()
```

```python
# t-test of volatile acidity on quality=3 vs. 8

x = data.loc[data['quality']==3, 'volatile acidity']
y = data.loc[data['quality']==8, 'volatile acidity']
ttest = stats.ttest_ind(x,y)

# show the data in a plot
plt.plot(np.random.randn(len(x))/30, x, 'o',
         np.random.randn(len(y))/30+1, y, 'o',
        )
plt.xlim(-1, 2)
plt.xticks([0,1],labels=['Qual 3','Qual 8'])
plt.title(f't={ttest[0]:.2f}, p={ttest[1]:.5f}')
plt.ylabel('volatile acidity')
plt.show()
```

```python
# show a histogram of quality scores

plt.hist(data.quality);
```

```python
data['quality'].value_counts().sort_index().plot(kind='bar', rot=0);
```

```python
# t-test of volatile acidity on quality=3 vs. 8

x = data.loc[(data['quality']==3) | (data['quality']==4), 'volatile acidity']
y = data.loc[(data['quality']==8) | (data['quality']==7), 'volatile acidity']
ttest = stats.ttest_ind(x,y)

# show the data in a plot
plt.plot(np.random.randn(len(x))/30, x, 'o',
         np.random.randn(len(y))/30+1, y, 'o',
         markeredgecolor = 'k'
        )
plt.xlim(-1, 2)
plt.xticks([0,1],labels=['Qual 3+4','Qual 7+8'])
plt.title(f't={ttest[0]:.2f}, p={ttest[1]:.5f}')
plt.ylabel('volatile acidity')
plt.show()
```
