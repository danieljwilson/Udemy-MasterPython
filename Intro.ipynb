```
import numpy as np
import matplotlib.pyplot as plt
```

```
N = 10000

# init x and y vectors
sx = np.zeros(N)
sy = np.zeros(N)

for i in range(1,N):
  # generate a random number {1,2,3}
  k = np.random.randint(1,4)

  # update the x and y points
  sx[i] = sx[i-1]/2 + k-1
  sy[i] = sy[i-1]/2
  if k==2:
    sy[i] += 2

plt.plot(sx,sy, 'k.', markersize=1)
plt.title('Sierpinski triangle')
plt.axis('off')
plt.show()
```

```
# OOP (classes)

# can have functions/methods attached to class

class Experiment(object):
  # constructor method
  def __init__(self,eType,dataN,status='ongoing'):
    self.expType = eType
    self.dataCount = dataN
    self.expStatus = status

  def increaseN(self,number):
    self.dataCount += number
  
  def changeExpStatus(self,status):
    self.expStatus = status

  def __str__(self):
    return 'Experiment type: ' + self.expType + ' has ' + str(self.dataCount) + ' data points.'
```

```
myexperiment = Experiment('Brain', 120)

print(myexperiment.dataCount)
myexperiment.increaseN(100)
print(myexperiment.dataCount)

print(myexperiment)
```

```
# Exercise
# create a class to store information about your board game hobby

class Boardgame(object):
  def __init__(self,playHistory,name='Mousetrap'):
    self.playHistory = playHistory
    self.playCount = len(playHistory)
    self.winCount = sum(i>0 for i in playHistory)
    self.name = name
    self.lastGame = playHistory[-1]
  
  def updatePlayCount(self,newGames):
    self.playHistory.extend(newGames)
    self.playCount = len(self.playHistory)
    self.winCount = sum(i>0 for i in self.playHistory)
    self.lastGame = self.playHistory[-1]

  def __str__(self):
    if self.lastGame > 0:
      last = 'win'
    else:
      last = 'loss'
    return 'Outcome of the last game was a '+ last + '!'
```

```
# create dot product function
# function checks that both inputs are numpy arrays
# function checks for same length of two input vectors
# returns dot product if pass, useful error message otherwise
```

```
def dotProduct(array1, array2):

  if not (isinstance(array1, np.ndarray) and isinstance(array2, np.ndarray)):
    raise Exception('Must be numpy array')
  if not len(array1) == len(array2):
    raise Exception('Must be the same length')
  
  return sum(array1*array2)
```

```
# Exercise: make a Poisson-like counter

# parameter lambda. positive number (10)
# start at 1, multiply by random number uniform([0,1])
# until target number is smaller than exp(-lambda)
# count number of iterations until target<exp(-lambda)
```

```
import numpy as np

def poissonCounter(lam):
  result = 1
  counter = 0
  target = np.exp(-lam)

  while result>target:
    result*=np.random.uniform()
    counter+=1

  return counter
```

```
p = np.zeros(100)

for i in range(100):
  p[i] = poissonCounter(10)

p
```

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```
n = 200
D = np.zeros((n,2))

D[:,0] = np.linspace(0,10,n) + np.random.randn(n)
D[:,1] = D[:,0]**2 + np.random.randn(n)*10
```

```
sns.jointplot(x=D[:,0], y=D[:,1])
plt.show()
```

```
df = pd.DataFrame(data=D, columns=['var1', 'var2'])
df
```

```
sns.jointplot(x = df.columns[0], y = df.columns[1], data=df)
plt.show()
```

```
sns.jointplot(x = df.columns[0], y = df.columns[1], data=df,
              kind='kde', color='purple')
plt.show()
```

```
x = np.linspace(-1,1,n)

y1 = x**2
y2 = np.sin(3*x)
y3 = np.exp(-10*x**2)

plt.plot(y1,y2,'o')
```

```
sns.scatterplot(x=y1, y=y2, hue=y3, legend=False, palette='Spectral')
plt.show()
```

```
# Exercise: regression plot of data df using seaborn

sns.regplot(x = df.columns[0], y = df.columns[1], data=df,
            color='green')

# can add title with matplotlib
plt.title(f'Regression of {df.columns[0]} on {df.columns[1]}')
plt.show()
```

```
# Exercise
# 1) create Hilbert matrix (10x10) H[i,j] = 1 / (i+j-1)
# 2) visualize the Hilbert matrix using seaborn!
```

```
x = 10
y = 10

H = np.zeros(shape=(x,y))

for i in range(1,x+1):
  for j in range(1,y+1):
    H[i-1,j-1] = 1/ (i+j-1)

H
```

```
import seaborn as sns

sns.heatmap(H)
plt.show()
```

```
sns.heatmap(H,vmin=0,vmax=.3)
plt.show()
```

```
# Sierpinski triangle OYO
# updated method for generating random numbers
rng = np.random.default_rng(12345)

n = 10000
D = np.zeros((n,2))

for i in range(1,n):
  k =rng.integers(low=1, high=4)
  D[i,0] = 0.5*D[i-1,0] + k-1
  D[i,1] = 0.5*D[i-1,1]
  if k==2:
    D[i,1]+=2

df = pd.DataFrame(data=D, columns=['x', 'y'])
```

```
import seaborn as sns
```

```
sns.scatterplot(data=df, x='x', y='y', markers=',')
plt.show()
```

```
plt.plot(df.x, df.y, marker=',', linestyle = 'None')
#plt.savefig('test.pdf')
plt.title('Sierpinski triangle')
plt.axis('off')
plt.show()
```
