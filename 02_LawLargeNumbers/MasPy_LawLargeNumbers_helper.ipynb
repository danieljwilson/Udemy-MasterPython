```python
# load modules
import numpy as np
import matplotlib.pyplot as plt
```

# Generate a population of random numbers

```python
# simulation parameters

population_size = 2.3e5

# 50 samples of 500 each
sample_size = 500
number_of_samples = 50 
```

```python
population_size
```

```python
# generate population distribution
f = np.logspace(np.log10(.001),np.log10(10), num=int(population_size))

# a population of numbers that decreases with increasing value
population = 1/f

# compute the true population mean
trueMean = np.mean(population)

# plot the data
skip = int(1e3)
plot_pop = population[::skip]
plt.plot(plot_pop, 'o')
plt.xlabel('sample')
plt.ylabel('Data value')
plt.show()
```

```python
# shuffle the data values
np.random.shuffle(plot_pop)
# plot the data
plt.plot(plot_pop, 'o')
plt.xlabel('sample')
plt.ylabel('Data value')
plt.show()
```

# Monte Carlo sampling

```python
# generate one random sample from the population

np.random.choice(population, 2)
```

```python
# Monte Carlo sampling for sample means

number_of_samples = 500
sample_means = np.zeros(number_of_samples) # initialize
sample_size = 50

for expi in range(number_of_samples):
    rand_sample = np.random.choice(population, size=sample_size)
    sample_means[expi] = np.mean(rand_sample)

  
```

```python
# plot the sample means, and a line showing the true mean
plt.plot(sample_means, 'ko', label='Sample means')
plt.hlines(np.mean(population), xmin=0, xmax=number_of_samples, color='r',
           linewidth=5, label='True mean')
plt.legend(loc='upper left')
plt.xlabel('Sample number')
plt.ylabel('Mean value')
plt.show()
```
